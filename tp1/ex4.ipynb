# Ejercicio 4

Variar la luminancia del pixel central en una de las dos imágenes (izquierda o derecha) hasta que perceptiblemente (por el ojo humano) posean el mismo brillo

```python
import matplotlib.pyplot as plt
from PIL import Image
```

```python
def horizontal_join_images(left_image: Image, right_image: Image) -> Image:
    joined = Image.new(mode="L", size=(left_image.width + right_image.width, max(left_image.height, right_image.height)))
    joined.paste(left_image, (0, 0))
    joined.paste(right_image, (left_image.width, 0))
    return joined
```

```python
LIGHT_SQUARE_COLOR = 223
DARK_SQUARE_COLOR = 63
SQUARE_SIZE = (9, 9)

PIXEL_COLOR = 127
PIXEL_SIZE = (3, 3)
```

```python
def build_image(left_delta: int, right_delta: int) -> Image:
    inside_square_left = Image.new(mode="L", size=PIXEL_SIZE, color=PIXEL_COLOR + left_delta)
    inside_square_right = Image.new(mode="L", size=PIXEL_SIZE, color=PIXEL_COLOR + right_delta)

    left_square = Image.new(mode="L", size=SQUARE_SIZE, color=DARK_SQUARE_COLOR)
    right_square = Image.new(mode="L", size=SQUARE_SIZE, color=LIGHT_SQUARE_COLOR)

    left_square.paste(inside_square_left, (3, 3))
    right_square.paste(inside_square_right, (3, 3))

    joined_squares = horizontal_join_images(left_square, right_square)
    return joined_squares
```

```python
def show_image(img: Image) -> None:
    plt.imshow(img, cmap="gray", vmin=0, vmax=255)
```

## Imágen sin variación

```python
show_image(build_image(0, 0))
```

## Variación de la luminancia del pixel central de la imágen de la derecha

```python
show_image(build_image(0, 80))
```

## Variación de la luminancia del pixel central de la imágen de la izquierda

```python
show_image(build_image(-55, 0))
```

## Conclusiones

- Al percibir la imágen original con el ojo humano, pese a tener la misma luminancia en el pixel central de ambas imágenes, aparenta más oscuro (menor luminancia) el de la derecha
- Variar el pixel central de la imágen derecha o el de la izquierda no es lo mismo. Variarla en el de la derecha tiene mayor impacto a menor variación comparado con la de la izquierda
