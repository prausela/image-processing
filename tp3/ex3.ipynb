```python
import numpy as np
import cv2
from pathlib import Path
from skimage.util import random_noise 
from scipy import signal
import skimage
```

```python
ASSETS_FOLDER_PATH = "./assets"
OUTPUT_FOLDER_PATH = "."
```

```python
Path(OUTPUT_FOLDER_PATH).mkdir(parents=True, exist_ok=True)
```

Cargamos la imagen

```python
eight = skimage.io.imread(fname=f"{ASSETS_FOLDER_PATH}/eight.tif")
```

```python
skimage.io.imshow(eight)
```

```python
sp_eight = random_noise(eight, mode='s&p', amount=0.02)
```

```python
skimage.io.imshow(sp_eight)
```

```python
average = np.ones((3, 3))
average = average / average.sum()
average
```

```python
avg_eight = cv2.filter2D(src=sp_eight, ddepth=-1, kernel=average) 
```

```python
skimage.io.imshow(avg_eight)
```

```python
med_eight = signal.medfilt2d(sp_eight, kernel_size=[3, 3])
```

```python
skimage.io.imshow(med_eight)
```
