```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
img = np.array(Image.open('./assets/moon.png').convert('L'))
height, width = img.shape
print(height, width)
```

```python
# (-1)^(x + y)
neg_one_pow = np.array([[-1**(x + y) for x in range(width)] for y in range(height)])

# image X (-1)^(x + y)
i_1 = img * neg_one_pow

# DFT
I_1 = np.fft.fft2(i_1)

# Conj
I_2 = np.conj(I_1)

# IDFT
i_2 = np.fft.ifft2(I_2)

# IDFT X (-1)^(x + y) 
final_img = i_2 * neg_one_pow
```

```python
plt.figure(figsize=(12,8)) 
ax = plt.subplot(1,2,1)
ax.imshow(img, cmap='gray')
ax.set_title('i(x,y)')

ax = plt.subplot(1,2,2)
ax.imshow(np.real(final_img), cmap='gray')
ax.set_title('i\'(x,y)')

plt.show()
```

La imagen final final_img es una versión reflejada y conjugada de la imagen original img. 
Específicamente, está invertida espacialmente y tiene los valores de intensidad reflejados en el sentido de conjugado complejo. Esto significa que si observas la imagen final, parecerá una versión "espejada" de la imagen original.
