# Ecualización de un histograma
La ecualización de un histograma es una transformación para mejorar el contraste en una imágen. Para ello se mapean los valores de intensidad de la imágen para que se encuentren distribuidos de manera más uniforme en el histograma.

Entonces, se debe elegir una transformación monótona creciente univaluada en el intervalo [0, L-1] (biyectiva) (L siendo el rango de intensidades).

Algunos de los usos posibles son:
- Para generar más contraste en la imágen
- Para "aclarar" imágenes muy oscuras. Al ecualizar, lo que se encuentra en la imágen es mucho más distinguible
- Para imágenes médicas. El contraste permite destacar lo que queremos del fondo. Por ejemplo en Rayos X, resonancias magneticas, etc.
- Mejorar la performance en tareas de computer vision
