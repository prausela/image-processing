```python
import numpy as np
from pathlib import Path
import skimage
```

```python
ASSETS_FOLDER_PATH = "./assets"
OUTPUT_FOLDER_PATH = "."
```

```python
Path(OUTPUT_FOLDER_PATH).mkdir(parents=True, exist_ok=True)
```

Cargamos la imagen

```python
lena = skimage.io.imread(fname=f"{ASSETS_FOLDER_PATH}/lena_gray.tif")
```

```python
skimage.io.imshow(lena)
```

```python
def negative(image: np.ndarray) -> np.ndarray:
    negative_image = image.copy()
    for n_row in range(image.shape[0]):
        for n_col in range(image.shape[1]):
            negative_image[n_row, n_col] = 255 - image[n_row, n_col]
    return negative_image
```

```python
negative_lena = negative(lena)
```

```python
skimage.io.imshow(negative_lena)
```
