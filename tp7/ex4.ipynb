```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./assets/Lincoln from Penny.tif')
plt.imshow(img)
plt.show()
```

```python
# Matriz de 1s de 5x5 para extraer los bordes de la img
B = np.ones((5, 5), np.uint8) 
eroded = cv2.erode(img, B, iterations=1)
plt.imshow(eroded)
plt.show()
```

```python
plt.imshow(img - eroded)
plt.show()
```
